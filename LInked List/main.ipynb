```python
def PalindromeList(head):

    def reverse(head):
        curr = head
        prev = None
        while curr :
            _next_ = curr.next
            curr.next = prev 
            prev = curr 
            curr = _next_

        return prev
    
    # find middle 
    slow = fast = head
    while fast.next and fast.next.next :
        slow = fast 
        fast = fast.next.next 

    # reverse 
    new_head = reverse(slow.next)

    # compare 
    first = head 
    second = new_head
    while second :
        if first.data != second.data :
            reverse(new_head)
            return False

        first = first.next
        second = second.next

    # reverse
    reverse(new_head)
    return True
```

> `Delete all occurrences of a key in DLL : ` 

```python
def delete_occurance(head , target):
    curr = head
    while curr :
        if curr.val == target :
            if curr.val == head :
                head = head.next 
            prev_node = curr.prev
            next_node = curr.next 
            if next_node : next_node.prev = prev_node
            if prev_node : prev_node.next = next_node
            curr = next_node
        else : 
            curr = curr.next 
    return head
```

> `Find Pairs with Given Sum in Doubly Linked List :`

```python
def find_tail(head):
    curr = head 
    while curr.next :
        curr = curr.next
    return curr

def find_paris(head, target):
    ans = []
    left = head 
    right = find_tail(head)
    while left < right :
        _sum_ = left.val + right.val 
        if _sum_ == target :
            ans.append([left.val , right.val ])
            left = left.next 
            right = right.prev
        elif _sum_ < target :
            left = left.next 
        else :
            right = right.prev
    return ans
```

> `Rotate a LL : `

```python
def rotateRight(self, head, k):
    if not head : return None
    
    tail = head 
    length = 1 # because we are stopping before Null
    # calculating the length of list
    while tail.next :
        length += 1
        tail = tail.next 
    
    # if the k is multiple of length of list then 
    # even rotating will give you same list so dont rotate
    if k%length == 0 : return head

    rotations_needed = k%length
    # combine total list back -> front 
    # for easy rotation 
    tail.next = head
    curr = head
    prev = None
    move = length - rotations_needed
    # start "move" from curr to end i.e to break the list to rotate
    # prev will be at the new tail 
    # curr will be at the new head 
    while move :    
        move -= 1
        prev = curr
        curr = curr.next
    prev.next = None
    head = curr
    return head
```

> `Remove Duplicates from DLL : `

```python
def removeDuplicates(self, head):
    if not head or not head.next : return head
    curr = head
    temp = head.next
    while temp :
        if curr.val == temp.val :
            curr.next = temp.next
            if temp.next : 
                temp.next.prev = temp.prev 
            temp = curr.next 
        else :
            curr = curr.next 
            temp = curr.next

    return head 
```

```python
def reverse(self ,head):
    prev = None 
    curr = head 
    while curr :
        _next_ = curr.next 
        curr.next = prev 
        prev = curr 
        curr = _next_
    return prev

def find_kth_pos(self, temp , k):
    while temp and k :
        temp = temp.next 
        k -=1 
    return temp

def reverseKGroup(self, head, k ):
    temp = head 
    while temp : 
        # find the kth position 
        kth_node = self.find_kth_pos(temp , k)
        if not kth_node : # this will execute when it is last node because it has no k value
            if prev_node : 
                prev_node.next = temp
            break # we came to end so stop program
        # seperate the k sized list to reverse it 
        new_node = kth_node.next
        kth_node.next = None 
        # reverse the list 
        self.reverse(temp)
        # this will execute only first time because 
        # temp and head are both at same location 
        if temp == head :
            head = kth_node
        else :
            # connect the prev list with the new reversed list 
            prev_node.next = kth_node
        
        # first place the prev_node on temp 
        prev_node = temp
        # then move the temp to new_node 
        temp = new_node

    return head
```

> `Flatten a LL : `

```python
class List_Node:
    def __init__(self, val=0, next=None, child=None):
        self.data = val
        self.next = next
        self.child = child


def flattenLinkedList(head: List_Node) -> List_Node:
    # Assume merge two list : O(n1+ n2) = O(m+m) = O(2m)
    # total : O(2NM)
    def merge_two_lists( l1 , l2):
        # Time : O(N1 + N2)
        dummy = List_Node(-1)
        res = dummy
        while l1 and l2 :
            if l1.data < l2.data :
                res.child = l1
                # move res pointer on answer list 
                res = res.child
                # move list pointer on list 
                l1 = l1.child
            else :
                res.child = l2
                # move res pointer on answer list 
                res = res.child
                # move list pointer on list 
                l2 = l2.child
        res.next = None

        res.child = l1 or l2 
        return dummy.child

    def func(head : List_Node):
        # Time : O(N)
        # Space : O(N)
        if not head or not head.next :
            return head

        merge_head = func(head.next)

        return merge_two_lists(merge_head , head)

    return func(head)
```
