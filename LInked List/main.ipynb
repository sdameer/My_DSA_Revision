```python
class Node :
    def __init__(self , data= None):
        self.data = data
        self.next = None
```

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def append(self , x):
        new_node = Node(x)
        if  self.head is None:
            self.head = new_node
            return

        curr = self.head
        while curr.next is not None:
            curr = curr.next
        curr.next = new_node

    def length(self):
        index = 0
        curr = self.head
        while curr is not None:
            curr = curr.next
            index = index + 1
        return index
        
    def display(self):
        array = []
        curr = self.head
        while curr is not None:
            array.append(curr.data)
            curr = curr.next
        return array
    
    def get(self , index):
        if index < 0 or index >= self.length():
            return "Index out of range"

        i = 0
        curr = self.head
        while curr is not None:
            if  i == index:
                return curr.data
            curr = curr.next
            i = i + 1
                        
    
    def delete(self , index):
        if index < 0 or index >= self.length():
            return "Index out of range"
        
        if index == 0:
            self.head = self.head.next
            return
        
        curr = self.head
        prev = None
        i = 0 

        while curr is not None:
            if i == index:
                prev.next = curr.next
                return 
            prev = curr
            curr = curr.next
            i = i + 1   
                




ll = LinkedList()


print("Empty:", ll.display())          


ll.append(10)
ll.append(20)
ll.append(30)
ll.append(40)

print("After append:", ll.display())   


print("Length:", ll.length())          


print("Index 0:", ll.get(0))           
print("Index 2:", ll.get(2))           
print("Index 3:", ll.get(3))           
print("Index 4:", ll.get(4))           


ll.delete(0)
print("After deleting index 0:", ll.display())   


ll.delete(1)
print("After deleting index 1:", ll.display())   


ll.delete(1)
print("After deleting last node:", ll.display()) 


ll.delete(0)
print("After deleting only node:", ll.display()) 


print(ll.delete(0))                   


ll.append(100)
ll.append(200)

print("New list:", ll.display())      


print(ll.get(-1))                     
print(ll.get(5))                      


print(ll.delete(-1))                  
print(ll.delete(5))                   
```

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def append_at_start(self, x):
        new_node = Node(x)

        new_node.next = self.head
        self.head = new_node

    def append_at_end(self, x):
        new_node = Node(x)
        curr = self.head
        while curr.next is not None :
            curr = curr.next

        curr.next = new_node      

    def append_at_index(self, idx , value):
        # need to insert the node at the idx 0 indexed LL
        if idx <0 or idx >= self.length():
            return "Index out of range"
        
        new_node = Node(value)
        prev = None
        curr = self.head
        i = 0 
        while i < idx :
            prev = curr
            curr = curr.next
            i = i + 1

        prev.next = new_node
        new_node.next = curr

    def append_after_value(self , value_1 ,value):
        new_node = Node(value)
        curr = self.head

        while curr is not None:
            if curr.data == value_1:
                new_node.next = curr.next
                curr.next = new_node
                return
            curr = curr.next        
                
        # we came here is means we traversed entire list not found the value_1
        return f"{value_1} not in Linked List"        
            
    def delete_at_start(self):
        self.head = self.head.next

    def delete_at_end(self):
        curr = self.head
        while curr.next.next is not None:
            curr = curr.next
        curr.next = None

    def delete_at_index(self, idx):
        if idx <0 or idx >= self.length():
            return "Index out of range"
        
        prev = None
        curr = self.head
        i = 0
        while i < idx:
            prev = curr
            curr = curr.next
            i +=1

        prev.next = curr.next
    
    def delete_at_value(self , value):
        curr = self.head
        while curr.next.data != value:
            curr = curr.next
        curr.next = curr.next.next
        
    def length(self):
        l = 0
        curr = self.head
        while curr is not None:
            l += 1
            curr = curr.next
            
        return l

    def get_value_with_index(self , idx):
        if idx<0 or idx>= self.length():
            return "Index out of range"
        
        curr = self.head
        i = 0
        while i < idx:
            curr = curr.next
            i = i + 1

        # for 0 indexed -> the curr will stop at val
        return curr.data
    
    def get_index_with_value(self, value):
        # for 0 indexed
        curr = self.head
        i = 0
        while curr.data != value:
            curr = curr.next
            i = i + 1
        return i

    def display(self):
        array = []
        curr = self.head
        while curr is not None:
            array.append(curr.data)
            curr = curr.next

        return array

      
```
