Floor and ceil in sorted array

```python
nums = [1, 3, 5, 7, 9 , 11]
target = 10

ceil = floor = len(nums)

# ceil (next element - upper bound)
low = 0
high = len(nums)-1
while low <= high:
    mid = (low+high)//2
    if nums[mid] >= target:
        ceil = mid
        high = mid - 1
    else:
        low = mid + 1

# floor (prev element - lower bound)
low = 0
high = len(nums)-1

while low <= high:
    mid = (low+high)//2
    if nums[mid] <= target:
        floor = mid
        low = mid + 1
    else:
        high = mid - 1

print([floor , ceil])
```

Find the Smallest Divisor Given a Threshold

```python
import math


def _sum(nums , mid : int)->int :
    num = 0
    for i in nums :
        num += math.ceil(i/mid)
    return num

def smallestDivisor(nums, threshold: int) -> int:
    my_min  = 0
    low = 1
    high = max(nums)
    while low<=high :
        mid = (low + high )//2
        if _sum(nums , mid) <= threshold:
            my_min = mid
            high = mid - 1
        else : 
            low = mid + 1

    return my_min


nums = [1,2,5,9]
threshold = 6
smallestDivisor(nums , threshold)
```

 Minimum Number of Days to Make m Bouquets

```python
class Solution:  
    def possible(self , nums , day , m  ,k):
        _sum = 0
        count = 0
        # begin from start of array and count all the days where flowers will bloomm
        for n in nums:
            if n <= day:
                count += 1
            else :
                _sum += math.ceil(count//k)
                count = 0

        # handle last case 
        _sum += math.ceil(count//k)

        # if we got all the bouqets 
        if _sum >= m :
            return True
        else :
            return False

    def minDays(self, arr: List[int], m: int, k: int) -> int:
        # iterate from the day where flowers start to bloom and the day where flowers stop
        # in the given arr in question
        for day in range(min(arr), max(arr)+1):
            if self.possible(arr , day , m , k) == True:
                return day
        return -1        

```

```python
bloomDay = [1,10,3,10,2]
m = 3
k = 1

def possible( arr , day , m , k):
    count = 0
    _sum = 0
    for a in arr :
        if a <= day :
            count += 1
        else :                  
            _sum += count // k
            count = 0
    _sum += (count // k)   
    return _sum

def minDays(arr, m: int, k: int) -> int:
    if m*k > len(arr) : return -1

    low = min(arr)
    high = max(arr)
    while low <= high :
        mid = (low + high)//2 # day
        if possible(arr , mid , m , k ) < m :
            low = mid + 1
        else :
            high = mid - 1

    return low

minDays(bloomDay , m, k)
```
