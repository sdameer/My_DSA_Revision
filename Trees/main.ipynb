```python
class TreeNode:
    def __init__(self, val, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def preorder(node):
    if node is None:
        return

    print(node.val, end=" ")
    preorder(node.left)
    preorder(node.right)


def inorder(node):
    if node is None:
        return

    inorder(node.left)
    print(node.val, end=" ")
    inorder(node.right)


def postorder(node):
    if node is None:
        return

    postorder(node.left)
    postorder(node.right)
    print(node.val, end=" ")

print("Preorder:")
preorder(root)

print("\nInorder:")
inorder(root)

print("\nPostorder:")
postorder(root)
```

> B F S

```python
# BFS :
def levelOrder(root):
    ans = []
    if root is None:return ans
    q = deque([root])
    while q:
        size = len(q)
        level = []
        for i in range(size):
            node = q.popleft()
            if node.left:q.append(node.left)
            if node.right:q.append(node.right)
            level.append(node.val)
        ans.append(level)
    return ans

levelOrder(root)
```

> D F S

```python
# DFS :
def dfs(node):
    if node is None:
        return

    print(node.val, end=" ")

    dfs(node.left)
    dfs(node.right)


root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)

dfs(root)
```

> Pre order traversal Iterative : 

Root -> Left -> Right

```python
from collections import deque

class TreeNode:
    def __init__(self, val, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


# preorder iterative traversal
stack = [root]
res = []
while stack :
    node = stack.pop()
    if node.right:
        stack.append(node.right)
    if node.left :
        stack.append(node.left)
    res.append(node.val)

print(res)
```

> In - Order :

Left -> Root -> Right

```python
# as traversal starts with left 
# so when there is a node then push to stack 
# update node with left 
# and when popping pop and push right to stack 
stack = []
node = root
res = []

while True:
    if node:
        stack.append(node)
        node = node.left
    else:   # NO node => we are at leaf node 
        if not stack: # NO stack => processed whole Tree
            break

        node = stack.pop()
        res.append(node.val)
        node = node.right

print(res)
```

> Post - Order :

Left -> Right -> Root

```python
stack = [root]
res = []

while stack:
    node = stack.pop()
    res.append(node.val)
    if node.left :
        stack.append(node.left)
    if node.right :
        stack.append(node.right)

print(res[::-1])
```

 > ` Maximum Depth in BT	`

```python
def height(node: TreeNode):
    if node is None:
        return 0

    left = height(node.left)
    right = height(node.right)

    return 1 + max(left, right)

height(root)
```

 > ` Check for balanced binary tree`

```python
def isBalanced(node : TreeNode):
    if node is None :
        return 0 
    
    left = isBalanced(node.right)
    right = isBalanced(node.left)

    if abs(left - right )> 1:
        return -1
    
    if left == -1 : return -1
    if right == -1 : return -1
    return 1 + max(left , right)


root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(2)
root.left.left.left = TreeNode(3)

isBalanced(root) != -1
```

 > ` Diameter of Binary Tree`


```python
dia = 0

def DiameterOfTree(root):
    global dia

    if root is None:
        return 0

    left = DiameterOfTree(root.left)
    right = DiameterOfTree(root.right)

    dia = max(dia, left + right)

    return max(left, right) + 1

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)

DiameterOfTree(root)
print(dia)
```

 > ` Check if two trees are identical or not`


```python
root_p = TreeNode(1)
root_p.left = TreeNode(2)
root_p.right = TreeNode(3)

root_q = TreeNode(1)
root_q.left = TreeNode(2)
root_q.right = TreeNode(3)

def SameTree(p : TreeNode, q : TreeNode):
    if not p and not q : # both null node !! OKAY
        return True
    if not p or not q : # a node is null 
        return False
    if p.val != q.val:
        return False
    L = SameTree(p.left , q.left)
    R = SameTree(p.right , q.right)
    return L and R

SameTree(root_p, root_q)
```
