```python
class TreeNode:
    def __init__(self, val, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def preorder(node):
    if node is None:
        return

    print(node.val, end=" ")
    preorder(node.left)
    preorder(node.right)


def inorder(node):
    if node is None:
        return

    inorder(node.left)
    print(node.val, end=" ")
    inorder(node.right)


def postorder(node):
    if node is None:
        return

    postorder(node.left)
    postorder(node.right)
    print(node.val, end=" ")

print("Preorder:")
preorder(root)

print("\nInorder:")
inorder(root)

print("\nPostorder:")
postorder(root)
```

> B F S

```python
# BFS :
def levelOrder(root):
    ans = []
    if root is None:return ans
    q = deque([root])
    while q:
        size = len(q)
        level = []
        for i in range(size):
            node = q.popleft()
            if node.left:q.append(node.left)
            if node.right:q.append(node.right)
            level.append(node.val)
        ans.append(level)
    return ans

levelOrder(root)
```

> D F S

```python
# DFS :
def dfs(node):
    if node is None:
        return

    print(node.val, end=" ")

    dfs(node.left)
    dfs(node.right)


root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)

dfs(root)
```

> Pre order traversal Iterative : 

Root -> Left -> Right

```python
from collections import deque

class TreeNode:
    def __init__(self, val, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


# preorder iterative traversal
stack = [root]
res = []
while stack :
    node = stack.pop()
    if node.right:
        stack.append(node.right)
    if node.left :
        stack.append(node.left)
    res.append(node.val)

print(res)
```

> In - Order :

Left -> Root -> Right

```python
# as traversal starts with left 
# so when there is a node then push to stack 
# update node with left 
# and when popping pop and push right to stack 
stack = []
node = root
res = []

while True:
    if node:
        stack.append(node)
        node = node.left
    else:   # NO node => we are at leaf node 
        if not stack: # NO stack => processed whole Tree
            break

        node = stack.pop()
        res.append(node.val)
        node = node.right

print(res)
```

> Post - Order :

Left -> Right -> Root

```python
stack = [root]
res = []

while stack:
    node = stack.pop()
    res.append(node.val)
    if node.left :
        stack.append(node.left)
    if node.right :
        stack.append(node.right)

print(res[::-1])
```

 > ` Maximum Depth in BT	`

```python
def height(node: TreeNode):
    if node is None:
        return 0

    left = height(node.left)
    right = height(node.right)

    return 1 + max(left, right)

height(root)
```

 > ` Check for balanced binary tree`

```python
def isBalanced(node : TreeNode):
    if node is None :
        return 0 
    
    left = isBalanced(node.right)
    right = isBalanced(node.left)

    if abs(left - right )> 1:
        return -1
    
    if left == -1 : return -1
    if right == -1 : return -1
    return 1 + max(left , right)


root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(2)
root.left.left.left = TreeNode(3)

isBalanced(root) != -1
```

 > ` Diameter of Binary Tree`


```python
dia = 0

def DiameterOfTree(root):
    global dia

    if root is None:
        return 0

    left = DiameterOfTree(root.left)
    right = DiameterOfTree(root.right)

    dia = max(dia, left + right)

    return max(left, right) + 1

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)

DiameterOfTree(root)
print(dia)
```

 > ` Maximum path sum`


```python
def maxPathSum(root):
    ans = float('-inf')

    def maxp(node):
        nonlocal ans
        if node is None : return 0
        # this code with add -ve numbers which will reduce the path 
        # leftsum = maxp(node.left)
        # rightsum = maxp(node.right)
        # always remove -ve by comparing with 0
        leftsum = max(0 , maxp(node.left))
        rightsum = max(0 , maxp(node.right))
        ans = max(ans , node.val + leftsum + rightsum)
        return node.val + max(leftsum , rightsum)


    maxp(root)
    return ans

maxPathSum(root)
```

 > ` Check if two trees are identical or not`


```python
root_p = TreeNode(1)
root_p.left = TreeNode(2)
root_p.right = TreeNode(3)

root_q = TreeNode(1)
root_q.left = TreeNode(2)
root_q.right = TreeNode(3)

def SameTree(p : TreeNode, q : TreeNode):
    if not p and not q : # both null node !! OKAY
        return True
    if not p or not q : # a node is null 
        return False
    if p.val != q.val:
        return False
    L = SameTree(p.left , q.left)
    R = SameTree(p.right , q.right)
    return L and R

SameTree(root_p, root_q)
```

 > ` Zig Zag or Spiral Traversal`

```python
def zigzag(root : TreeNode):
    if root is None:
        return []
    
    q = deque([root])
    res = []
    level_index = 1
    while q :
        q_len = len(q)
        level = []
        for _ in range(q_len):
            node = q.popleft()
            level.append(node.val)
            if node.left :q.append(node.left)
            if node.right :q.append(node.right)
        if level_index%2==0:
            level.reverse()
        level_index += 1
        res.append(level)

    return res


zigzag(root)
```

 > ` Right Side View :`

```python
def rightSideView(root):
    ans = []
    level = 0
    def dfs(node, level):
        nonlocal ans 
        if node is None : return 0
        if len(ans) == level : ans.append(node.val)
        dfs(node.right , level+1)
        dfs(node.left , level+1)


        
    dfs(root , level)
    return ans

rightSideView(root)
# note : " level " is not being updated globally. Each recursive call gets its own copy of the level value.        
```

 > `Symmetric Tree : `

```python
def isSymmetric( root):
    def dfs(left , right):
        if not left and not right : return True
        if not left or not right : return False
        
        return (left.val == right.val and 
                dfs(left.left , right.right) and 
                dfs(left.right, right.left) )
    return dfs(root.left,root.right)

isSymmetric(root)
```

 > `Boundary level Traversal : `

```python
def boundary(root):
    leaves = []
    left_nodes = []
    right_nodes = []

    def leaf(node):
        if node is None:
            return

        if node != root and not node.left and not node.right:
            leaves.append(node.val)

        leaf(node.left)
        leaf(node.right)

    def left(node):
        if not node or (not node.left and not node.right):
            return

        left_nodes.append(node.val)

        left(node.left if node.left else node.right)

    def right(node):
        if not node or (not node.left and not node.right):
            return

        right_nodes.append(node.val)

        right(node.right if node.right else node.left)

    leaf(root)
    left(root.left)
    right(root.right)

    return [root.val] + left_nodes + leaves + right_nodes[::-1]

boundary(root)
```

> `Vertical Traversal : `

```python
ans = defaultdict(lambda: defaultdict(list))
q = deque([(root, 0, 0)])
res = []
while q:
    node, x, y = q.popleft()
    ans[x][y].append(node.val)

    if node.left:
        q.append((node.left, x-1, y + 1))
    if node.right:
        q.append((node.right, x + 1, y + 1))

for x in sorted(ans):
    temp = []
    for y in sorted(ans[x]):        
        temp.extend(ans[x][y])
    res.append(temp)
print(res)
```

> `Top View :`

```python
ans = {}
q = deque([(root , 0)])
while q:
    node , vert = q.popleft()
    if vert not in ans:
        ans[vert] = node.val
    if node.left :
        q.append((node.left, vert-1))
    if node.right:
        q.append((node.right , vert+1))

print([ans[i] for i in sorted(ans)])
```

> `Bottom View :`

```python
ans = {}
q = deque([(root , 0)])
while q:
    node , vert = q.popleft()
    
    ans[vert] = node.val
    
    if node.left :
        q.append((node.left, vert-1))
    if node.right:
        q.append((node.right , vert+1))

print([ans[i] for i in sorted(ans)])
```

> `Root to the Node : `

```python
arr = []

def get_path(node, x):
    if not node:
        return False
    
    # Assume the current node may be part of the answer,
    # so add it to the path.
    arr.append(node.val)

    # Target found.
    if node.val == x:
        return True

    # Search in both subtrees.
    # If either subtree finds the target, propagate True upward.
    left = get_path(node.left, x)
    right = get_path(node.right, x)

    if left or right:
        return True

    # Reaching here means:
    # 1. Current node is not None.
    # 2. Current node itself is not the target.
    # 3. Neither the left subtree nor the right subtree contains the target.
    # Therefore, this node is not part of the root-to-target path,
    # so undo our earlier choice and remove it.
    arr.pop()

    # Tell the parent that the target was not found
    # in the subtree rooted at the current node.
    return False

get_path(root , 7)
print(arr)
```

> `Lowest Common Ancestor : `

```python
def LCA(node, p , q):
    if node is None:
        return None
    if node.val == p or node.val == q:
        return node
    
    left = LCA(node.left , p , q)
    right = LCA(node.right , p , q)

    if left is None : return right
    elif right is None : return left
    else : return node

LCA(root ,2 , 3).val
```
