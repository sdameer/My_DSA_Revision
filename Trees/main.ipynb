````python
class TreeNode:
    def __init__(self, val, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def preorder(node):
    if node is None:
        return

    print(node.val, end=" ")
    preorder(node.left)
    preorder(node.right)


def inorder(node):
    if node is None:
        return

    inorder(node.left)
    print(node.val, end=" ")
    inorder(node.right)


def postorder(node):
    if node is None:
        return

    postorder(node.left)
    postorder(node.right)
    print(node.val, end=" ")

print("Preorder:")
preorder(root)

print("\nInorder:")
inorder(root)

print("\nPostorder:")
postorder(root)
````

> B F S

````python
# BFS :
def levelOrder(root):
    ans = []
    if root is None:return ans
    q = deque([root])
    while q:
        size = len(q)
        level = []
        for i in range(size):
            node = q.popleft()
            if node.left:q.append(node.left)
            if node.right:q.append(node.right)
            level.append(node.val)
        ans.append(level)
    return ans

levelOrder(root)
````

````python
def height(node: TreeNode):
    if node is None:
        return 0

    left = height(node.left)
    right = height(node.right)

    return 1 + max(left, right)

height(root)
````

 > ` Check for balanced binary tree`

````python
def isBalanced(node : TreeNode):
    if node is None :
        return 0 
    
    left = isBalanced(node.right)
    right = isBalanced(node.left)

    if abs(left - right )> 1:
        return -1
    
    if left == -1 : return -1
    if right == -1 : return -1
    return 1 + max(left , right)


root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(2)
root.left.left.left = TreeNode(3)

isBalanced(root) != -1
````

 > ` Diameter of Binary Tree`


````python
dia = 0

def DiameterOfTree(root):
    global dia

    if root is None:
        return 0

    left = DiameterOfTree(root.left)
    right = DiameterOfTree(root.right)

    dia = max(dia, left + right)

    return max(left, right) + 1

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)

DiameterOfTree(root)
print(dia)
````

 > ` Maximum path sum`


````python
def maxPathSum(root):
    ans = float('-inf')

    def maxp(node):
        nonlocal ans
        if node is None : return 0
        # this code with add -ve numbers which will reduce the path 
        # leftsum = maxp(node.left)
        # rightsum = maxp(node.right)
        # always remove -ve by comparing with 0
        leftsum = max(0 , maxp(node.left))
        rightsum = max(0 , maxp(node.right))
        ans = max(ans , node.val + leftsum + rightsum)
        return node.val + max(leftsum , rightsum)


    maxp(root)
    return ans

maxPathSum(root)
````

 > ` Zig Zag or Spiral Traversal`

````python
def zigzag(root : TreeNode):
    if root is None:
        return []
    
    q = deque([root])
    res = []
    level_index = 1
    while q :
        q_len = len(q)
        level = []
        for _ in range(q_len):
            node = q.popleft()
            level.append(node.val)
            if node.left :q.append(node.left)
            if node.right :q.append(node.right)
        if level_index%2==0:
            level.reverse()
        level_index += 1
        res.append(level)

    return res


zigzag(root)
````

 > ` Right Side View :`

````python
def rightSideView(root):
    ans = []
    level = 0
    def dfs(node, level):
        nonlocal ans 
        if node is None : return 0
        if len(ans) == level : ans.append(node.val)
        dfs(node.right , level+1)
        dfs(node.left , level+1)


        
    dfs(root , level)
    return ans

rightSideView(root)
# note : " level " is not being updated globally. Each recursive call gets its own copy of the level value.        
````

 > `Symmetric Tree : `

````python
def isSymmetric( root):
    def dfs(left , right):
        if not left and not right : return True
        if not left or not right : return False
        
        return (left.val == right.val and 
                dfs(left.left , right.right) and 
                dfs(left.right, right.left) )
    return dfs(root.left,root.right)

isSymmetric(root)
````

 > `Boundary level Traversal : `

````python
def boundary(root):
    leaves = []
    left_nodes = []
    right_nodes = []

    def leaf(node):
        if node is None:
            return

        if node != root and not node.left and not node.right:
            leaves.append(node.val)

        leaf(node.left)
        leaf(node.right)

    def left(node):
        if not node or (not node.left and not node.right):
            return

        left_nodes.append(node.val)

        left(node.left if node.left else node.right)

    def right(node):
        if not node or (not node.left and not node.right):
            return

        right_nodes.append(node.val)

        right(node.right if node.right else node.left)

    leaf(root)
    left(root.left)
    right(root.right)

    return [root.val] + left_nodes + leaves + right_nodes[::-1]

boundary(root)
````

> `Root to the Node : `

````python
arr = []

def get_path(node, x):
    if not node:
        return False
    
    # Assume the current node may be part of the answer,
    # so add it to the path.
    arr.append(node.val)

    # Target found.
    if node.val == x:
        return True

    # Search in both subtrees.
    # If either subtree finds the target, propagate True upward.
    left = get_path(node.left, x)
    right = get_path(node.right, x)

    if left or right:
        return True

    # Reaching here means:
    # 1. Current node is not None.
    # 2. Current node itself is not the target.
    # 3. Neither the left subtree nor the right subtree contains the target.
    # Therefore, this node is not part of the root-to-target path,
    # so undo our earlier choice and remove it.
    arr.pop()

    # Tell the parent that the target was not found
    # in the subtree rooted at the current node.
    return False

get_path(root , 7)
print(arr)
````

> `Maximum Width of BT : `

````python
def widthOfBinaryTree(root):
    q = deque([(root, 0)])        
    ans = 0
    while q :
        start = q[0][1]
        end = q[-1][1]
        q_len = len(q)
        for _ in range(q_len):
            node , index = q.popleft()
            index = index - start
            if node.left :
                q.append((node.left , 2*index + 1))
            if node.right :
                q.append((node.right , 2*index+2))
        ans = max(ans , end - start + 1)
    return ans

widthOfBinaryTree(root)
````

> ` Children Sum Property in Binary Tree : `

````python
def changeTree(root):
    if not root:
        return

    child_sum = 0

    if root.left:
        child_sum += root.left.val
    if root.right:
        child_sum += root.right.val

    # Top-down adjustment
    if child_sum < root.val:
        if root.left:
            root.left.val = root.val
        if root.right:
            root.right.val = root.val
    else:
        root.val = child_sum

    changeTree(root.left)
    changeTree(root.right)

    # Bottom-up correction
    total = 0
    if root.left:
        total += root.left.val
    if root.right:
        total += root.right.val

    if root.left or root.right:
        root.val = total

changeTree(root)
````

````python
def distanceK(root, target, k):
    parent = {} # parent : node
    def inorder(node):
        if not node : return 
        if node.left :
            parent[node.left] = node
        inorder(node.left)
        if node.right :
            parent[node.right] = node
        inorder(node.right)

    inorder(root)
    def bfs(root, k):
        q = deque([root])
        visited = {root}
        while q :
            if k == 0:
                return [node.val for node in q]
            q_len = len(q)
            for _ in range(q_len):

                node = q.popleft()

                if node.left and node.left not in visited:
                    q.append(node.left)
                    visited.add(node.left)

                if node.right and node.right not in visited:
                    q.append(node.right)
                    visited.add(node.right)

                if node in parent and parent[node] not in visited:
                    q.append(parent[node])
                    visited.add(parent[node])

            k = k - 1

        return []
    
    return bfs(target , k)
````

> `Count total nodes in a complete BT : `

````python
def countNodes(root) :
    def findleft(node):
        h = 0
        while (node):
            h += 1
            node = node.left
        return h

    def findright(node):
        h = 0
        while node :
            h += 1
            node = node.right
        return node
    if not root : 
        return 0
    left = findleft(root.left)
    right = findright(root.right)
    if left == right :
        return 2^left - 1

    return 1 + countNodes(root.left) + countNodes(root.right)

countNodes(root)
````

> `Construct Binary Tree from InOrder and PostOrder`

````python
def buildTree(inorder, postorder):
    _index = {}
    for i, j in enumerate(inorder):
        _index[j] = i

    def helper(l, r):
        if l > r:
            return None
        root = TreeNode(postorder.pop())
        idx = _index.get(root.val)
        root.right = helper(idx+1, r)
        root.left = helper(l, idx-1)
        return root

    return helper(0, len(inorder) - 1)


buildTree(
    [9, 3, 20, 15, 7],
    [9, 15, 7, 20,  3]
).val
````

> `Serialize and Deserialize Binary Tree :`

````python
def serializer(root):
    if not root :
        return ""
    q = deque([root])
    ans = []
    while q :
        node = q.popleft()
        if node :
            ans.append(str(node.val))
            q.append(node.left)
            q.append(node.right)
        else :
            ans.append("#")

    return ",".join(ans)

serializer(root)
````

````python
def deserialize(data):
    values = data.split(",")
    root = TreeNode(int(values[0]))
    q = deque([root])
    i = 1
    while q :
        node = q.popleft()
        if values[i]!="#":
            node.left = TreeNode(int(values[i]))
            q.append(node.left)
        i = i + 1
        if values[i] != "#":
            node.right = TreeNode(int(values[i]))
            q.append(node.right)
        i = i + 1

    return root

deserialize('1,2,3,4,5,6,#,#,#,#,#,#,#').val
````

> `Morris Traversal :`

````python
def FindPredecessor(node):
    temp = node.left
    while temp.right and temp.right != node:
        temp = temp.right    
    return temp

ans = []
def morris(node):
    curr = node
    while curr!= None :
        if not curr.left :
            ans.append(curr.val)
            curr = curr.right
        else :
            pre = FindPredecessor(curr)
            if not pre.right :
                pre.right = curr
                curr = curr.left
            else :
                pre.right = None
                ans.append(curr.val)
                curr = curr.right                
    return ans

morris(root)
````

````python
def FindPredecessor(node):
    temp = node.left
    while temp.right:
        temp = temp.right
    return temp

def flatten(node):
    curr = node
    while curr!= None :
        if curr.left:
            pre = FindPredecessor(curr)

            pre.right = curr.right
            curr.right = curr.left 
            curr.left = None
    curr = curr.left
````

> `Search in BST : `

````python
def searchBST(root, val):
    while root is not None and root.val != val:
        root = root.left if val < root.val else root.right
    return root

searchBST(root , val=1).val
````

> `Ceil in BST : `

````python
def findceil(root, key):
    ceil = -1
    while root :
        if root.val == key:
            ceil = root.val
            return ceil
        
        # val >= key ✅
        if root.val > key:
            ceil = root.val
            root = root.left
        else:
            root = root.right

    return ceil

findceil(root , key=2)
````

> `Floor in BST : `

````python
def findfloor(root , key):
    floor = -1
    while root :
        if root.val == key :
            floor = root.val
            return floor
        
        # val <= key ✅
        if root.val < key:
            floor = root.val
            root  = root.right 
        else :
            root = root.left
    return root

findceil(root , key=2)
````

> `Delete a Node : ` 

To DELETE a node we do these things :
    
    search the node
    
    delete the node
    
    adjust the B.S.T

````python
def findpred(node):
    temp = node.left
    while temp.right :
        temp = temp.right
    return temp

def helper(node):
    if not node.left : return node.right
    if not node.right : return node.left
    root = node
    right_most_chlid = findpred(node)
    left_chlid = node.left
    right_most_chlid.left = left_chlid

    return root.left

def delete(node, key):
    curr = node
    while curr :
        if key < curr.val :
            if curr.left and curr.left.val == key:
                curr.left = helper(curr.left)
                break
            else :
                curr = curr.left
        else :
            if curr.right and curr.right.val == key :
                curr.right = helper(curr.right)
                break
            else :
                curr = curr.right



print("Before : " , levelOrder(root))
delete(root , 8)
print("After  : ",levelOrder(root))
````

> `Construct BST from preorder : ` 

````python
def bst_preorder(preorder):
    i = 0 

    def build(bound):
        nonlocal i
        if i == len(preorder) or preorder[i] > bound:
            return None

        root = TreeNode(preorder[i])
        i = i + 1

        root.left = build(root.val)
        root.right = build(bound)

    return build(float('inf'))

bst_preorder([8 , 7 , 1, 10 , 12])
````
