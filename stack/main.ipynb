```python
def NGR(nums):
    """
    when the stack contains any elements which is smaller
    than the current element then pop it 
    stack should contain only the right large elements
    always assign the `res` with the top most elements of stack 
    because we are storing the right most element in the stack
    """
    n = len(nums)
    res = [-1] * n
    stack = []
    for i in range(n - 1 , - 1, -1):        
        while stack and stack[-1] <= nums[i]:
                stack.pop()
        if stack :
            res[i] = stack[-1]
        
        stack.append(nums[i])

    return res
```

```python
def NSR(nums):
    n = len(nums)
    res = [-1] * n
    stack = []

    for i in range(n - 1 , -1 , -1):
        while stack and stack[-1] > nums[i]:
             stack.pop()
        if stack:
             res[i] = stack[-1]

        stack.append(nums[i])

    return res
```

```python
def NSL(nums):
    n = len(nums)
    res = [-1] * n
    stack = []
    for i in range(n):
        while stack and stack[-1] > nums[i]:
            stack.pop()
        if stack:
            res[i] = stack[-1]
        stack.append(nums[i])

    return res
```

```python
def NGL(nums):
    n = len(nums)
    stack = []
    res = [-1] * n
    for i in range(n):
        while stack and stack[-1] <= nums[i]:
            stack.pop()
        
        if stack:
            res[i] = stack[-1]

        stack.append(nums[i])

    return res
```

```python
nums = [1 , 3 , 2, 4]


print(NGR(nums))
print(NSR(nums))
print(NSL(nums))
print(NGL(nums))
```
