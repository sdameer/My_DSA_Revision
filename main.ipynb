 `Binary Search`

```python
nums = [2, 3, 4, 5, 6]
target = 5

low = 0
high = len(nums) - 1
while low<= high:
    mid = (low+high)//2
    if nums[mid] == target :
        print(mid)
    if target <= nums[mid]:
        high = mid -1
    else :
        low = mid + 1
```
